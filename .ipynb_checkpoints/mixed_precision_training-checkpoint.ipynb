This notebook is designed to show the effectiveness of mixed precision training on reducing GPU RAM usage, as well as how it can speed up the training process itself

```python
import tensorflow as tf
import numpy as np
import timeit
import matplotlib.pyplot as plt
from tensorflow.keras import mixed_precision
```

# Data (Images)

Create images using random noise, and create binary labels randomly for use with the images in a classification model

```python
# create dummy images based on random data
SEED = 12
tf.random.set_seed(SEED);
total_images = 800
input_shape = (total_images, 256, 256, 3) # (batch, height, width, channels)
images = tf.random.uniform(input_shape, minval=0.0, maxval=1.0, seed=SEED, dtype=tf.float32)
labels = np.random.choice([0, 1], size=(total_images,), p=[0.5,0.5])
```

```python
# plot an example image
image_example = images[0]
plt.imshow(image_example);
```

# Neuralnet Function

A multilayer convolutional neural network for testing

```python
def neural_net(images=images, labels=labels, batch_size=50, epochs=10):
    # layers
    layer1 = tf.keras.layers.Conv2D(128,2)
    layer2 = tf.keras.layers.Conv2D(128,1)
    layer3 = tf.keras.layers.Conv2D(128,1)
    layer4 = tf.keras.layers.Flatten()
    layer5 = tf.keras.layers.Dense(1)
    output_layer = tf.keras.layers.Activation('sigmoid',dtype=tf.float32)

    # print data type information
    print(f'GENRAL AND LAYER DTYPE INFO:\n')
    print(f'Global policy: {mixed_precision.global_policy()}')
    print(f'Compute dtype: {mixed_precision.global_policy().compute_dtype}')
    print(f'Variable dtype: {mixed_precision.global_policy().variable_dtype}')
    print(f'inputs dtype: {images.dtype}')
    print(f'layer1 dtype policy: {layer1.dtype_policy}')
    print(f'layer2 dtype policy: {layer2.dtype_policy}')
    print(f'layer3 dtype policy: {layer3.dtype_policy}')
    print(f'layer4 dtype policy: {layer4.dtype_policy}')
    print(f'layer5 dtype policy: {layer5.dtype_policy}')
    print(f'output dtype policy: {output_layer.dtype_policy}\n')
    print(f'RUN MODEL:\n')

    # model
    inputs = tf.keras.layers.Input(shape=input_shape[1:])
    x = layer1(inputs)
    x = layer2(x)
    x = layer3(x)
    x = layer4(x)
    x = layer5(x)
    outputs = output_layer(x)
    model = tf.keras.Model(inputs,outputs)

    # compile
    model.compile(optimizer=tf.keras.optimizers.Adam(),
                  loss='binary_crossentropy',
                  metrics=['accuracy'])
    
    history = model.fit(images, labels, batch_size=batch_size, epochs=epochs)

    return model.summary()
```

# Print Function

This provides GPU RAM information and is used before and after each run to see how much GPU RAM was used during the test.

```python
def print_ram_info():
    if tf.config.list_physical_devices('GPU'):
        # Returns a dict in the form {'current': <current mem usage>,
        #                             'peak': <peak mem usage>}
        memory = tf.config.experimental.get_memory_info('GPU:0')
        print(f'\nRAM INFO:\n\nCurrent: {memory["current"]/1e+09:.2f} GB, Peak: {memory["peak"]/1e+09:.2f} GB, USED MEMORY FOR RUN: {(memory["peak"] - memory["current"])/1e+09:.2f} GB\n')
```

# Setup code for the timeit module. 

These blocks of code are run only once by timeit, not in the loop

```python
setup_normal = '''
import tensorflow as tf
from tensorflow.keras import mixed_precision
mixed_precision.set_global_policy('float32')
from __main__ import neural_net
tf.config.experimental.reset_memory_stats('GPU:0')
tf.keras.backend.clear_session()
'''

setup_mixed_precision = '''
import tensorflow as tf
from tensorflow.keras import mixed_precision
mixed_precision.set_global_policy('mixed_float16')
from __main__ import neural_net
tf.config.experimental.reset_memory_stats('GPU:0')
tf.keras.backend.clear_session()
'''
```

# Mixed Precision Test

*   Input is float32
*   Neuralnet changes what it can to float16 automatically

```python
print('MIXED PRECISION RUN')
print('-------------------')
if tf.config.list_physical_devices('GPU'):
    tf.config.experimental.reset_memory_stats('GPU:0')
print_ram_info()
mixed_time = timeit.timeit('neural_net()', number=1, setup=setup_mixed_precision)
print_ram_info()
print(f'TIME TO COMPLETE RUN: {mixed_time:.2f}')
```

# Normal Precision Test

*   Input is float32
*   Neuralnet runs with float32



```python
print('NORMAL PRECISION RUN')
print('--------------------')
if tf.config.list_physical_devices('GPU'):
    tf.config.experimental.reset_memory_stats('GPU:0')
print_ram_info()
normal_time = timeit.timeit('neural_net()', number=1, setup=setup_normal)
print_ram_info()
print(f'TIME TO COMPLETE RUN: {normal_time:.2f}')
```
